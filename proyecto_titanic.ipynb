```python
df_train = df_train.drop(columns=['PassengerId','Ticket','Cabin','Name'], axis=1)
```

### Preprocesamiento

```python
from sklearn.compose import make_column_selector
from sklearn.pipeline import make_pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder,StandardScaler,OrdinalEncoder
from sklearn.compose import make_column_transformer
```

```python
cat_colum_nominal = ['Embarked']
cat_colum_ordinal = ['Sex']
num_selector = make_column_selector(dtype_include='number',dtype_exclude='object')
```

```python
# Transformadores nominal
one_hot_encoder = OneHotEncoder()
# Transformadores ordinal
ordinal_encoder = OrdinalEncoder()
# Standar scaler
standard_scaler = StandardScaler()
# Standar Scaler (Trasformador de scala)
imputer_cat = SimpleImputer(strategy='most_frequent')
imputer_num = SimpleImputer(strategy='mean')
```

```python
col_transformer = make_column_transformer(
    (make_pipeline(imputer_num, standard_scaler), num_selector),
    (make_pipeline(imputer_cat, one_hot_encoder), cat_colum_nominal),
    (make_pipeline(ordinal_encoder), cat_colum_ordinal),
    remainder='passthrough'
)
col_transformer
```

```python
X_train = df_train.drop(columns=['Survived'])
y_train = df_train['Survived']
```

### Entrenamiento de datos con el algoritos DecisionTreeClassifier

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.pipeline import make_pipeline

# Define el clasificador
classifier = DecisionTreeClassifier(max_depth = 10,
                             random_state = 0)

# Crea el pipeline combinando el transformador de columnas y el clasificador
pipeline = make_pipeline(col_transformer, classifier)

# Entrena el modelo usando el conjunto de entrenamiento
pipeline.fit(X_train, y_train)
```

```python
# Preprocesar las características del conjunto de prueba utilizando col_transformer
# Hacer predicciones sobre el conjunto de prueba preprocesado
predictions = pipeline.predict(df_test)
```
